### Theory

When we want a computer to understand how similar two pieces of text are, we need a way to **measure that similarity mathematically**.

Classical ML gives us simple, interpretable, and computationally efficient tools to do this **without training a neural network**. The three methods below form a natural progression — each one addresses a real weakness in the previous one.
## #Method 1: Jaccard Similarity (JS)


Jaccard Similarity measures how much two **sets of words** overlap. It treats every document as a **bag of unique words** (no counts, no weights, just present or absent) and checks what fraction of the combined vocabulary is shared.



### Formula

$$
JS(A, B) = \frac{|A \cap B|}{|A \cup B|}
$$

Where:
- $A$ and $B$ are the **sets of unique words** in document A and document B
- $A \cap B$ = words that appear in **both** documents (intersection)
- $A \cup B$ = all words that appear in **either** document (union)
- Result is always between **0** (no overlap) and **1** (identical word sets)



### Simple Example 


| **Paper 1** | `"Vision AI Vision"` -> unique words: `{vision, AI}` -> Size = 2 

| **Paper 2** | `"NLP AI"` -> unique words: `{NLP, AI}` -> Size = 2 

**Step-by-step:**

- Intersection = `{AI}` -> size = **1** (only AI is shared)
- Union = `{vision, AI, NLP}` -> size = **3**

$$
JS(\text{Paper1, Paper2}) = \frac{1}{3} \approx 0.33
$$

 Even though "Vision" appears twice in Paper 1, it is only counted once in the set. Jaccard works on *unique* words only.



### Limitation of Jaccard Similarity

> **Problem: It ignores how many times a word appears (word frequency).**

- Repeated words are ignored — `"Vision AI Vision"` and `"Vision AI"` produce the same set.
- Common words like `the`, `is`, `for` are weighted the same as rare meaningful words like `transformer`.
- **All words are treated as equally important** — which is not how language works.

This pushed researchers toward a method that could represent documents as **numerical vectors** and actually count word frequency.



## Method 2: Cosine Similarity


Instead of comparing word sets, we represent each document as a **numerical vector** — one number per word in the shared vocabulary. Then we measure the **angle** between two vectors. If the angle is small (vectors point in the same direction), the documents are similar.



### Formula

$$
CS(A, B) = \frac{A \cdot B}{|A| \times |B|}
$$

Where:
- $A \cdot B$ = **dot product**: multiply corresponding elements and sum
- $|A|$ = **magnitude (length)** of vector A = $\sqrt{\sum a_i^2}$
- $|B|$ = **magnitude (length)** of vector B = $\sqrt{\sum b_i^2}$
- Result is between **0** (completely different) and **1** (identical direction)



### Simple Example (from the board)

**Shared vocabulary** = `[vision, AI, NLP]` (3 dimensions)

| vision | AI | NLP | Vector |

 **Paper 1:** Vision AI Vision | 2 | 1 | 0 | `[2, 1, 0]` 

 **Paper 2:** NLP AI | 0 | 1 | 1 | `[0, 1, 1]` 

**Dot product:**

$$A \cdot B = (2 \times 0) + (1 \times 1) + (0 \times 1) = 0 + 1 + 0 = 1$$

**Magnitudes:**

$$|A| = \sqrt{2^2 + 1^2 + 0^2} = \sqrt{5} \approx 2.236$$

$$|B| = \sqrt{0^2 + 1^2 + 1^2} = \sqrt{2} \approx 1.414$$

**Cosine Similarity:**

$$CS = \frac{1}{2.236 \times 1.414} = \frac{1}{3.162} \approx 0.316$$

Vision appearing **twice** in Paper 1 is now captured in the vector as `2`. Word frequency matters here.

**How it's used with KNN:**
Once we have cosine similarity scores, we use **K-Nearest Neighbours (KNN)** to retrieve the K most similar documents to a query.



### Limitation of Cosine Similarity (with raw counts)

> **Problem: All words are still treated as equally important.**

- Common words like `the`, `is`, `for`, `with` appear in nearly every document.
- Their high counts **inflate similarity** between documents that only share filler words.
- A paper about Transformers for NLP and one about CNNs for CV might both use `for` many times and appear artificially similar.

In short: **frequency is captured, but not *importance*.**

This pushed researchers toward giving each word a weight based on how rare or common it is across the entire collection.

## Method 3: TF-IDF (Term Frequency - Inverse Document Frequency)

### Core Idea

TF-IDF assigns a **weight** to every word in every document. The weight is **high** if the word appears often in *that specific document* but **rarely** across *all documents*. Common filler words get low weights automatically.

It has **two components** multiplied together:

---

### Formula

$$\text{TF-IDF}(t, d) = TF(t, d) \times IDF(t)$$

**Part 1 - Term Frequency (TF):** How often does word $t$ appear in document $d$?

$$TF(t, d) = \log(1 + \text{count}(t, d))$$

> The **log** dampens very high counts. A word appearing 100 times should not be 100 times more important than one appearing once.

**Part 2 - Inverse Document Frequency (IDF):** How rare is word $t$ across all $N$ documents?

$$IDF(t) = \log\left(\frac{1 + N}{1 + df(t)}\right) + 1$$

Where:
- $N$ = total number of documents in the collection
- $df(t)$ = number of documents that contain word $t$
- The `+1` prevents zero IDF for very common terms (smooth IDF)

---


### Limitation of TF-IDF

> **Problem: It still treats each word as independent — it has no understanding of meaning.**

- `car` and `automobile` are treated as **completely different words**, even though they mean the same thing.
- `bank (finance)` and `bank (river)` are treated as **the same word**, even though they mean different things.
- TF-IDF cannot understand **context**, **synonyms**, or **word relationships**.

This limitation eventually motivated the development of **word embeddings** (Word2Vec, GloVe) and then **Transformers** (BERT, GPT), which learn dense vector representations capturing semantic meaning.

### 4) K-Nearest Neighbors (KNN)


K-Nearest Neighbors (KNN) is a **supervised machine learning algorithm** used for both **classification** and **regression**. It classifies a data point based on how its neighbors are classified.



#### Key Idea
- Similar data points are located close to each other.
- Distance or similarity measures are used to identify neighbors.



#### Steps of KNN Algorithm
1. Choose the number of neighbors (**K**)
2. Calculate distance/similarity between query point and all training data
3. Select the **K nearest neighbors**
4. Perform:
   - **Classification:** Majority voting
   - **Regression:** Average of values



#### Distance Measures

#### Euclidean Distance
$$
d = \sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2}
$$

#### Cosine Similarity
$$
CS = \frac{A \cdot B}{|A| |B|}
$$



#### Example
- Paper 1 → [2, 1, 0]  
- Paper 2 → [0, 1, 1]  

Steps:
- Compute similarity
- Choose K nearest documents
- Assign label based on majority



### Choosing K
- Small K → Sensitive to noise  
- Large K → More stable but less precise  
- Common values: **K = 3, 5, 7**



#### Advantages
- Simple and easy to implement  
- No training phase (lazy learning)  
- Works well for small datasets  



#### Limitations
- Slow for large datasets  
- Sensitive to irrelevant features  
- Requires feature scaling  



#### Connection with Cosine Similarity
- Instead of Euclidean distance, KNN can use **cosine similarity**
- Higher cosine similarity → closer neighbor
- Useful in **text classification problems**

```python


import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# -----------------------------
# 1. Synthetic Research Dataset
# -----------------------------
paper_titles = [
    ("ResNet-3D: Residual Grids for Spatial Volumetric Analysis", "CV", "CNNs", "Dr_Y_LeCun"),
    ("ViT-Scale: Scaling Vision Transformers for Dense Prediction", "CV", "Transformers", "Dr_A_Vaswani"),
    ("SimCLR-V3: Contrastive Frameworks for Self-Supervised Vision", "CV", "CL", "Dr_G_Hinton"),
    ("Masked Autoencoders Are Scalable Vision Visualizers", "CV", "Transformers", "Dr_Y_LeCun"),
    ("NeRF-Graph: Neural Radiance Fields on Graph Topologies", "CV", "GNNs", "Dr_J_Leskovec"),
    ("Real-Time Semantic Segmentation via Dilated Spatial Kernels", "CV", "CNNs", "Dr_Y_LeCun"),

    ("BERT-Large: Pre-training of Deep Bidirectional Transformers", "NLP", "Transformers", "Dr_A_Vaswani"),
    ("GPT-Next: Autoregressive Language Modeling at Scale", "NLP", "Transformers", "Dr_A_Vaswani"),
    ("Contrastive Sentence Embeddings via Semantic Invariance", "NLP", "CL", "Dr_Y_Bengio"),
    ("Long-Short Sequence Parsing via Linear Attention Windows", "NLP", "Transformers", "Dr_A_Vaswani"),
    ("Text-Graph Recurrent Transformers for Structured Document Analysis", "NLP", "GNNs", "Dr_J_Leskovec"),

    ("GCN-V2: Scalable Graph Convolutional Networks via Node Sampling", "Graph_AI", "GNNs", "Dr_J_Leskovec"),
    ("Graph Attention Networks with Multi-Head Structural Alignment", "Graph_AI", "Transformers", "Dr_A_Vaswani"),
    ("Self-Supervised Graph Contrastive Learning via Subgraph Masking", "Graph_AI", "CL", "Dr_Y_Bengio"),
    ("Temporal Graph Networks for Dynamic Relational Interaction Stream", "Graph_AI", "GNNs", "Dr_J_Leskovec"),
]

# -----------------------------
# 2. Convert papers → text docs
# -----------------------------
paper_texts = [
    f"{title} {domain} {method} {author}"
    for title, domain, method, author in paper_titles
]

N = len(paper_texts)  # total number of documents

# 3. TF-IDF
# Step 3a: Get raw term counts (TF numerator)
count_vectorizer = CountVectorizer()
count_matrix = count_vectorizer.fit_transform(paper_texts).toarray()  # shape: (N, V)
vocab = count_vectorizer.get_feature_names_out()

# Step 3b: Term Frequency (TF) — log normalization
#   TF(t, d) = log(1 + count(t, d))
#   Dampens the effect of high raw counts
tf_matrix = np.log1p(count_matrix)  # log(1 + count), shape: (N, V)

# Step 3c: Inverse Document Frequency (IDF) — log-smoothed
#   DF(t) = number of docs containing term t
#   IDF(t) = log((1 + N) / (1 + DF(t))) + 1   ← sklearn's smooth IDF formula
#   The +1 outside prevents zero IDF for very common terms
df = np.count_nonzero(count_matrix, axis=0)          # shape: (V,)
idf_vector = np.log((1 + N) / (1 + df)) + 1          # shape: (V,)

# Step 3d: TF-IDF = TF × IDF (element-wise broadcast)
tfidf_matrix = tf_matrix * idf_vector                 # shape: (N, V)

# Step 3e: L2-normalize each document row so dot product = cosine similarity
row_norms = np.linalg.norm(tfidf_matrix, axis=1, keepdims=True)
row_norms[row_norms == 0] = 1                         # avoid division by zero
tfidf_matrix_norm = tfidf_matrix / row_norms          # shape: (N, V)

# 4. Cosine Similarity 

# Cosine similarity between two vectors u, v:
#   cos(u, v) = (u · v) / (||u|| * ||v||)
# Since tfidf_matrix_norm rows are already L2-normalized:
#   cos(u_norm, v_norm) = u_norm · v_norm  (just a dot product)
cosine_sim_matrix = cosine_similarity(tfidf_matrix_norm)  # shape: (N, N)

# 5. Jaccard Similarity — 
def jaccard_similarity(a: str, b: str) -> float:
    """
    Jaccard(A, B) = |A ∩ B| / |A ∪ B|
    Uses the unique word-sets of two strings.
    """
    set_a = set(a.lower().split())
    set_b = set(b.lower().split())
    if not set_a and not set_b:
        return 0.0
    return len(set_a & set_b) / len(set_a | set_b)


# 6. Recommendation Function
def recommend(paper_index: int, top_k: int = 5) -> None:
    query_title = paper_titles[paper_index][0]
    print(f"\n{'='*70}")
    print(f"Query Paper:  {query_title}")
    print(f"{'='*70}")

    cosine_scores  = []
    tfidf_scores   = []
    jaccard_scores = []

    for i in range(N):
        if i == paper_index:
            continue

        # ── Cosine Similarity (from the normalised TF-IDF vectors) ──
        cos_score = cosine_sim_matrix[paper_index][i]
        cosine_scores.append((i, cos_score))

        # ── TF-IDF Score: dot product of raw (non-normalised) rows ──
        #    Measures how much shared weighted vocabulary the two docs have
        #    without the length-normalisation step that Cosine applies.
        raw_tfidf_score = np.dot(tfidf_matrix[paper_index], tfidf_matrix[i])
        tfidf_scores.append((i, raw_tfidf_score))

        # ── Jaccard Similarity (independent of TF-IDF) ──
        jac_score = jaccard_similarity(paper_texts[paper_index], paper_texts[i])
        jaccard_scores.append((i, jac_score))

    # Sort each ranking independently
    cosine_scores.sort(key=lambda x: x[1], reverse=True)
    tfidf_scores.sort(key=lambda x: x[1], reverse=True)
    jaccard_scores.sort(key=lambda x: x[1], reverse=True)

    # ── Print: Cosine Similarity ──
    print(f"\n{'─'*70}")
    print(" Cosine Similarity  ")
    print(f"{'─'*70}")
    for i, score in cosine_scores[:top_k]:
        print(f"  {paper_titles[i][0]:<58} | {score:.4f}")

    # ── Print: TF-IDF Raw Score ──
    print(f"\n{'─'*70}")
    print(" TF-IDF Score  ")
    print(f"{'─'*70}")
    for i, score in tfidf_scores[:top_k]:
        print(f"  {paper_titles[i][0]:<58} | {score:.4f}")

    # ── Print: Jaccard ──
    print(f"\n{'─'*70}")
    print("Jaccard Similarity  ")
    
    print(f"{'─'*70}")
    for i, score in jaccard_scores[:top_k]:
        print(f"  {paper_titles[i][0]:<58} | {score:.4f}")

# 7. Run example

recommend(0)
```

## Conclusion

Classical ML techniques like **Jaccard**, **Cosine Similarity**, and **TF-IDF** form the foundation of text similarity and document retrieval. Each method builds directly on the shortcomings of the previous:

- **Jaccard** is simple but blind to how often words appear.
- **Cosine Similarity** captures frequency but cannot distinguish important words from filler words.
- **TF-IDF** fixes importance weighting but still has no understanding of language semantics or meaning.

Understanding these classical methods is essential because:
1. They are still widely used in production search and recommendation systems.
2. They provide the **intuition and baseline** against which modern deep learning methods like Transformers and BERT are compared.
3. They are **interpretable** — you can always explain why a document was recommended.
